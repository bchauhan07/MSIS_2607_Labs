```python
""" function to print sequence of numbers of lenght provided by user. Print keyword "Fizz" for multiples of 3,
"Buzz" for multiples of 5 and "FizzBuzz for multiples of 15 """
def fizzbuzz():
    # take in a user input
    while True:
        mynumber = input("Choose a number: ")
        if mynumber.isnumeric():
            break
        else:
            print("That was not a number, please try again.")
            # input validation
    
    mynumber = int(mynumber)
    for i in range(1, mynumber+1):
        if i % 15==0:
            print("FizzBuzz")
        elif i % 3==0:
            print("Fizz")
        elif i % 5==0:
            print("Buzz")
        else:
            print(i)

fizzbuzz()
```

```python
# function to print prime number upto length entered by user

def primenumber():
    # take user input
    while True:
        number = input("Choose a number: ")
        if number.isnumeric():
          break
        else:
            print ("That was not a number, please try again.")
            # input validation
    number = int(number)
    print ("Prime numbers upto %d is: " %number)
    for num in range(number):
        if num >1:
            for i in range(2,num):
                # check if number not divisible by any other number other than itself and 1
                if num%i == 0:
                    break
            else:
                print (num)

primenumber()
            
```
