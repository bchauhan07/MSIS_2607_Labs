```python
# Lab 1
"""@  Team: 1. Vishakha Srivastava (vsrivastava@scu.edu)
            2. BInny Chauhan (bchauhan@scu.edu)
            3. Lisa Patel (lpatel@scu.edu)
            4. Pratiksha Raval (praval@scu.edu)
            5. Hithesh Sekhar ()
""" 
    
    
"""Function to print Fibonacci series of length provided by user """
def fibonacci_print():
    # Take input from user
    while True:
        n = input("Choose a number:")
        if not n.isnumeric():
            print("That was not a number. Please try again.")
        else:
            break
            # Input validation
    var1 = 0
    var2 = 1
    n = int(n)
    print ("Fibonacci sequence:")
    if n == 1:
        print (var1)
    elif n == 2:
        print (var1)
        print (var2)
    else:
        print (var1)
        print (var2)
        for i in range(2,int(n)):
            var3 = var1+var2
            var1=var2
            var2=var3
            print(var3)


""" Function to print sequence of numbers of lenght provided by user. Print keyword "Fizz" for multiples of 3,
"Buzz" for multiples of 5 and "FizzBuzz for multiples of 15 """
def fizzbuzz():
    # Take in a user input
    while True:
        mynumber = input("Choose a number: ")
        if mynumber.isnumeric():
            break
        else:
            print("That was not a number, please try again.")
            # input validation
    
    mynumber = int(mynumber)
    for i in range(1, mynumber+1):
        if i % 15==0:
            print("FizzBuzz")
        elif i % 3==0:
            print("Fizz")
        elif i % 5==0:
            print("Buzz")
        else:
            print(i)


"""Function to print prime number upto length entered by user"""
def primenumber():
    # Take user input
    while True:
        number = input("Choose a number: ")
        if number.isnumeric():
          break
        else:
            print ("That was not a number, please try again.")
            # input validation
    number = int(number)
    print ("Prime numbers upto %d : " %number)
    for num in range(number):
        if num >1:
            for i in range(2,num):
                # check if number not divisible by any other number other than itself and 1
                if num%i == 0:
                    break
            else:
                print (num)


""" Function to print random number list between -50 and 50. The lenght of the list will be given by user"""
import random
def randomnumber():
    # take input from user
    while True:
        number = input("Choose a number: ")
        if number.isnumeric():
            break
        else:
            print("That was not a number, please try again.")
            # input validation
    number = int(number)
    lst = []
    for i in range(number):
        lst.append(random.randrange(-50,51))
        # Append list with random numbers between -50 and 50
    print (lst)
    
#randomnumber()

def exit_menu():
    quit()

"""User Menu"""
def menu():
    usermenu = {
        '1':fibonacci_print,
        '2':fizzbuzz,
        '3':primenumber,
        '4':randomnumber,
        '5':exit_menu
    }
    userchoice = ''
    while (userchoice != '5'):
        # print user menu    
        for key in usermenu.keys():
            value = usermenu.get(key)
            print(str(key) + ". " + value.__name__)
        # take input from user
        userchoice = input("Choose the operation to be performed: ")    
        if userchoice in usermenu.keys():
            usermenu[userchoice]()
            print ("")
        else:
            print("Invalid input. Please choose an option from menu ")
menu() 
```
